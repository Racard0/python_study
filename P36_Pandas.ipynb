# P40 聚合函数

```python
var1_name
```

```python
group2 = vgsales1[var1_name].groupby('Year')
```

```python
vgsales1[var1_name]
```

```python
# agg - 聚合
# agg方法灵活,能够对分组对象进行相同的聚合，还可以选择不同的聚合方
# 法
group2.agg([np.mean,np.sum])
```

```python
group2.agg([np.mean,np.sum]).loc[[1980,1981],('NA_Sales',
                                              ['mean','sum'])]
```

```python
group2.agg({'NA_Sales':np.sum,'EU_Sales':np.mean})
```

```python
group2.agg({'NA_Sales':[np.median,np.sum],\
            'EU_Sales':[np.mean,np.sum]})
```

```python
# 自定义函数
def DoubleSUM(data):
    s = data.sum() * 2
    return s
group2.agg({'EU_Sales':DoubleSUM})
# 相当于
# group2.sum()['EU_Sales'] * 2
```

```python
group2.sum()['EU_Sales']
```

```python
group2.sum()['EU_Sales'] * 2
```

```python
vgsales1.agg({'EU_Sales':DoubleSUM})
```

```python
# 相当于
# 全部求和再 *2
vgsales1['EU_Sales'].sum() * 2
```

```python
var_name
```

```python
# apply - apply可以进行聚合计算，还可以按行计算
vgsales1[var_name].apply(np.sum,axis=0)
```
