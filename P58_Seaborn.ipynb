箱线图

```python
Titanic=pd.read_csv('titanic_train.csv')
```

```python
Titanic
```

```python
Titanic.dropna(subset=['Age'],inplace=True)
# 删除空值
```

```python
Titanic
```

```python
sns.boxplot(y='Age',
           data=Titanic,
           showmeans=True,
           color='steelblue',
            width=0.3,
           flierprops={'marker':'o','markerfacecolor':'indianred','markersize':5},
           #异常值                   填充色                        大小
           meanprops={'marker':'D','markersize':8},
            #均值               D:菱形
           medianprops={'linestyle':'--','color':'orange'})
           #中位数
plt.ylabel('年龄')
plt.show()
```

```python
sec_building=pd.read_excel('sec_buildings.xlsx')
```

```python
sec_building
```

```python
group_region=sec_building.groupby('region') #分组聚合
```

```python
avg_price=group_region.agg({'price_unit':np.mean}) #基于列的聚合操作
```

```python
avg_price.sort_values('price_unit',ascending=False) #ascending:是否升序
```

```python
avg_price=avg_price.sort_values('price_unit',ascending=False)
```

```python
avg_price.index
```

```python
plt.figure(figsize=(12,9))
# 分组箱线图
sns.boxplot(x='region',
           y='price_unit',
           data=sec_building,
           order=avg_price.index, #order:坐标轴
           showmeans=True,
           color='steelblue',
           width=0.3,
           flierprops={'marker':'o','markerfacecolor':'indianred','markersize':5},
           meanprops={'marker':'D','markerfacecolor':'black','markersize':3,},
           medianprops={'linestyle':'--','color':'orange'})
plt.title('不同地区二手房价格对比')
plt.xticks(rotation=45)
plt.ylabel('单位价格(元)')
plt.show()
```

直方图

```python
Titanic
```

```python
Age_Male=Titanic[Titanic.Sex == 'male']['Age'] #男性
Age_Female=Titanic[Titanic.Sex == 'female']['Age'] #女性
```

```python
from scipy.stats import norm
```

```python
sns.distplot(Age_Male,
            bins=30, #bins:直方条形个数
            kde=False, #kde:是否绘制核密度图
            hist_kws={'color':'g'}, #hist_kws:传递直方图相关参数
            norm_hist=True, #norm_hist=是否将频数改成频率
            label='男性')
sns.distplot(Age_Male,
             hist=False, #hist:是否绘制直方图
            bins=30,
            kde=False,
             fit=norm, #fit:是否绘制概率密度图
            fit_kws={'color':'g','linestyle':'--'}, 
             #fit_kws：传递概率密度曲线相关参数
            norm_hist=True,
            label='男性正态分布图')

sns.distplot(Age_Female,
            bins=30, 
            kde=False, 
            hist_kws={'color':'y'},
            norm_hist=True, 
            label='女性')
sns.distplot(Age_Female,
             hist=False,
            bins=30,
            kde=False,
             fit=norm,
            fit_kws={'color':'y','linestyle':'--'},
            norm_hist=True,
            label='女性正态分布图')
plt.legend(frameon=True)
plt.xlabel('年龄')
plt.ylabel('频率')
plt.show()
```

```python
income_data=pd.read_csv('income_data.csv')
```

```python
income_data
```

```python
income_data['year']=range(1994,2014,1) #增加一列
```

```python
income_data
```

```python
# 背景设置
sns.set(style='darkgrid',
       context='notebook') 

sns.lineplot(x='year',
            y='y',
            data=income_data,
            lw=2,
            color='red')
plt.xticks(range(1994,2014,1),range(1994,2014,1),rotation=45)
#          设置刻度           设置刻度标签
plt.show()
```

回归图

```python
sns.lmplot(x='x3',
          y='y',
          data=income_data,
          legend_out=False,
          markers='o',
          fit_reg=True,
           scatter_kws={'s':30,'facecolor':'red'}, #scatter_kws:散点设置
          aspect=1.3, #aspect:宽度
          height=8) #height:高度
plt.show()
```
