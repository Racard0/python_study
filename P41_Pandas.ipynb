```python
df = pd.read_csv('vgsales.csv',dtype={'Year':float})
```

```python
df
```

```python
# pivot_table - 数据透视表
# aggfunc - 聚合函数
# margins - 是否需要总计
# fill-value - 缺失值
# index - 行分组键
# columns - 列分组键
# values - 分组字段，只能为数值型变量
pd.pivot_table(data = df,index='Year',columns='Genre',values='Global_Sales',
              aggfunc=np.mean,margins=True,margins_name='总平均值',
              fill_value=0)
# 更多的可以参考“Pandas运用第三节.pdf”
```

```python
# crosstab - 交叉表
pd.crosstab(index=df['Platform'],columns=df['Genre'],margins=True)
```

```python
pd.crosstab(index=df['Platform'],columns=df['Genre'],margins=True,
           values=df['NA_Sales'],aggfunc=np.sum)
```

```python
# normalize - 查看占比
pd.crosstab(index=df['Platform'],columns=df['Genre'],margins=True,
           normalize='all') #all - 总数占比
                            #index - 行百分比
                            #columns - 列百分比
```

```python
df2 = pd.read_excel('Online_Retail.xlsx')
```

```python
df2.shape
```

```python
# 抽样
# sample - 样品
data = df2.sample(n=50000,replace=False)
```

```python
data
```

```python
data.dtypes
```

```python
data['date'] = data['InvoiceDate'].dt.date
```

```python
data['date']
```

```python
# 总价=单价×数量
data['total_price']=data[['Quantity','UnitPrice']].apply(np.prod,axis=1)
```

```python
data['total_price']
```

```python
# 按日期汇总
grouped_data = data.groupby('date').sum()
```

```python
grouped_data
```

```python
grouped_data.index
# dtype='object'
```

```python
# 行索引转化为日期格式
grouped_data.index = pd.to_datetime(grouped_data.index)
```

```python
grouped_data.index
# dtype='datetime64[ns]'
```

```python
# 变化率
# pct - 每百分
grouped_data['总价变化率'] = grouped_data['total_price'].pct_change()
```

```python
grouped_data
```

```python
# rolling() - 单位平均
grouped_data['SMA_10']=grouped_data['total_price'].rolling(10).mean()
# rolling(5) - 每五日计算平均
grouped_data['SMA_10']
```

```python
grouped_data['SMA_5']=grouped_data['total_price'].rolling(5).mean()
```

```python
# plot() - 绘制
grouped_data[['SMA_5','SMA_10']].plot()
```

```python
grouped_data.head(5)
```

```python
# shift(1) - 向下平移一个单位
# -1就向上平移
grouped_data['total_price_1'] = grouped_data['total_price'].shift(1)
```

```python
grouped_data['total_price_1']
```

```python
grouped_data['total_price_-1'] = grouped_data['total_price'].shift(-1)
```

```python
grouped_data['total_price_-1']
```

```python
grouped_data[['total_price_1','total_price','total_price_-1']]
```

```python
grouped_data['total_price'].pct_change()
```

```python
# pct_change()，相当于
# 变化率：当天-前一天/前一天
(grouped_data['total_price'] - grouped_data['total_price_1']) / grouped_data['total_price_1']
# 同上(grouped_data['total_price'] - grouped_data['total_price'].shift(1)) / grouped_data['total_price'].shift(1)
```

```python
# diff() - 一阶差分
# 即df - df.shift(1)
grouped_data['total_price'].diff()
```

```python
# 即也相当于
grouped_data['total_price'].diff() / grouped_data['total_price'].shift(1)
```
