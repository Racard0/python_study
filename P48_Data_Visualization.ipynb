# P48 数据可视化 matplotlib绘图基础

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# 创建一个等差数列0-100，1000个
x = np.linspace(0,100,1000)
```

```python
x
```

```python
y=np.sin(x)
```

```python
plt.plot?
```

```python
# x轴y轴
# c - 颜色，lw - 线条宽度
# ls - 线条风格('-','--','-.',':')
plt.plot(x,y,c='red',lw=2,ls=':')
```

```python
# marker:线条上点的形状
# markersize:点的大小
# markeredgecolor:点的边框色
# markerfacecolor:点的填充色
plt.plot(x,y,ls='-.',marker='o',markersize=3,c='red',
        markeredgecolor='blue',markerfacecolor='black',
        label='x and y')
plt.legend()
plt.show()
```

```python
# 解决中文乱码
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
# plt.legend:展示图例
# upper center:上中心
# center:中心
x1 = np.linspace(0,10,100)
y1 = np.sin(x1)
plt.plot(x1,y1,ls=':',lw = 3,label='x1 和 y1')
plt.legend(loc = 'center') #legend：图例
plt.show()
```

```python
#             起始，终止，步长
x2 = np.arange(0,1.1,0.01)
x2
```

```python
y2 = x2**2
```

```python
# figure:画布比例
# facecolor:背景色,默认white
# dpi:分辨率
plt.figure(figsize=(6.4,4.84),dpi=100,facecolor='blue')
plt.title('picture') #标题
plt.xlabel('x2') #x轴名称
plt.ylabel('y2') #y轴名称
plt.xlim([0,1])  #x轴范围
plt.ylim([0,1])  #y轴范围
plt.xticks([0,0.2,0.4,0.6,0.8,1]) #x轴的刻度
plt.yticks([0,0.2,0.4,0.6,0.8,1]) #y轴的刻度
plt.plot(x2,y2,label='y=x^2')
plt.legend(loc='best') #默认'best',自动选择
# savefig:保存图片
plt.savefig(r'E:\云开见明\云开见明Python数据分析师课程数据（仅与云开见明Python数据分析师配套）\第四章\test1 _18_picture.png')
plt.show()
```

# P49 简单图形绘制

```python
import os 
os.chdir('E:\云开见明\云开见明Python数据分析师课程数据（仅与云开见明Python数据分析师配套）\第四章')
```

```python
data=pd.read_excel('Prod_Trade.xlsx')
```

```python
data
```

```python
data.groupby('Region').mean()
```

```python
data_1=data.groupby('Region').mean()['Sales']
```

```python
data_1
```

```python
data_1.index
```

```python
x_data = data_1.values
y_labels = data_1.index
```

```python
x_data
```

```python
# explode:突出显示
explode = [0,0.1,0,0,0,0.5]
colors = ['red','blue','yellow','green','darkred','purple']
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #解决中文乱码
plt.rcParams['axes.unicode_minus'] = False #正负号显示
```

###### 饼图

```python
# autopct：保留百分比小数
# pctdistance：百分比标签与圆心距离
# labeldistance：各扇形标签距离与圆心距离，1.1在圆外
# startangle：饼图初始摆放角度
# radius：半径
# countercolock：是否逆时针
# wedgeprops：饼图内外边界
# textprops：饼图中文本属性
plt.pie(x=x_data,
        explode=explode,
        labels=y_labels,
        colors=colors,
        autopct='%.1f%%',
        pctdistance=0.5,
        labeldistance=1.1,
        startangle=120,
        radius=1.2,
        counterclock=False,
        wedgeprops={'linewidth':1.5,'edgecolor':'green'},
        textprops={'fontsize':10,'color':'black'})
plt.title('订单金额分布图',pad=30) #pad:标题位置
plt.show()
```

###### 条形图

```python
x_data
```

```python
y_labels
```

```python
plt.bar(x=range(0,len(x_data)),
        height=x_data,
        align='center', #align:x轴对齐位置
        color='y',      #yellow
        tick_label=y_labels) #tick_label:条形图刻度标签
plt.xlabel('地区',labelpad=30) #labelpad:标签位置
plt.ylabel('销量',labelpad=50)
plt.title('不同地方销量',pad=50)
plt.show()
```

###### 直方图

```python
plt.hist(x=data['Sales'],
         bins=20, #bins:条形个数
         color='r',
        edgecolor='black',
        density=True) #desity:以频数方式展现
plt.title('销量分布图')
plt.show()
```

###### 散点图

```python
sec_buildings=pd.read_excel('sec_buildings.xlsx')
```

```python
sec_buildings
```

```python
plt.scatter(x=sec_buildings['size'],
           y=sec_buildings['price'],
           color='steelblue',  #steelblue：刚蓝色
           marker='o',         #marker:散点图形状
           s=100)              #s:散点大小
plt.xlabel('房子大小',fontsize=40) #fontsize:字体大小
plt.ylabel('房屋总价',fontsize=40)
plt.title('房子大小和房屋总价关系',fontsize=40)
plt.show()
```

#  P51 图形基本设置

```python
plt.plot(data_1.index,data_1.values,label='不同地区销量')
plt.legend(loc='upper left',fontsize=10,frameon=False) 
#frameon:图例是否有边框

fig=plt.gcf() #gcf:返回该图像并设置为fig
fig.set_size_inches(10,5) #设置图像大小
plt.show()
```

```python
plt.plot(data_1.index,data_1.values,label='不同地区销量')
plt.legend(loc='upper left',fontsize=10,frameon=True) 
# 网格线
# grid:网格
plt.grid(ls='--',c='darkblue')
plt.show()
```

```python
data_1.index
```

```python
plt.plot(data_1.index,data_1.values,label='不同地区销量')
plt.legend(loc='upper left',fontsize=10,frameon=False) 
# 水平参考线
plt.axhline(y=1700,c='red',ls='--',lw=2) #横线，默认为0
plt.axvline(x=4,c='y',ls='--',lw=2) #竖线平行于y轴，与x轴垂直
plt.show()
```

```python
data_1.min()
```

```python
plt.plot(data_1.index,data_1.values,label='不同地区销量')
plt.legend(loc='upper left',fontsize=10,frameon=False)
# axvspan绘制一个参考区域
# alpha,透明度
plt.axvspan(xmin=4,xmax=5,facecolor='r',alpha=0.1) #竖状
plt.axhspan(ymin=1666,ymax=1700,facecolor='y',alpha=0.1) #横坐标
plt.show()
```
