# P34 条件查询和增删改查

```python
order[['dishes_id','dishes_name']][order['order_id']==458]
```

```python
# amount - 金额
# & - and
# | - or
order[['order_id','dishes_name']][(order['order_id'] == 458)\
                                 &(order['amounts']>50)]
```

```python
order[['order_id','dishes_name']][(order['order_id'] == 458)\
                                 |(order['amounts']>50)]
```

```python
# ~ - 取反
order[['order_id','dishes_name']][~(order['order_id']==458)]
```

```python
order['amounts'].between(10,30,inclusive=True)
# between
```

```python
# 30-50元
order[['detail_id','dishes_name','amounts']][order['amounts'].between(30,50,inclusive = True)]
```

```python
# isin - 判断元素是否存在
order['dishes_name'].isin(['蒜蓉生蚝'])
```

```python
order[['detail_id','dishes_name','amounts']][order['dishes_name'].isin(['蒜蓉生蚝','青椒肉丝'])]
```

```python
# 使用字符串的方法
# str.contains
order['dishes_name'].str.contains('蒜蓉')
```

```python
order[['detail_id','dishes_name','amounts']][order['dishes_name'].str.contains('蒜蓉')]
```
